```python
## creating the Class Kettle
class Kettle(object):
    
    power_source = "electricity"
    
    ## method is the function in the class
    ## method 1 - function 1
    def __init__(self, make, price):
        #print("The self is Called :---")
        print("The Kettle is Made of the Company :--", make)
        print("The Kettle is of the Price :--", price)
        self.make = make
        self.price = price
        self.on = False
    
    ## method 2 - function 2    
    def switch_on(self):
        print("The Self is Called :--")
        self.on = True
        
kenwood = Kettle("Kenwood", 12.75)
hamilton = Kettle("Hamilton", 15.99)

kenwood.power = 1.5  ## added new power attribute to the kenwood
hamilton.switch_on() ## switched on the hamilton switch by calling the second method in the class


print("*"*80)

print(Kettle.power_source)
print(kenwood.power_source)
print(hamilton.power_source)

print("*"*80)

print(Kettle.__dict__)
print(kenwood.__dict__)
print(hamilton.__dict__)

print("Switch the Power Source to Atomic")
Kettle.power_source = "Atomic Power"
print("*"*80)

print(Kettle.power_source)
print(kenwood.power_source)
print(hamilton.power_source)
print("*"*80)

print(Kettle.__dict__)
print(kenwood.__dict__)
print(hamilton.__dict__)

print("Switch the Kenwood Power Source to Gas Power.")
kenwood.power_source = "Gas Power"
print("*"*80)

print(Kettle.power_source)
print(kenwood.power_source)
print(hamilton.power_source)

print("*"*80)

print(Kettle.__dict__)
print(kenwood.__dict__)
print(hamilton.__dict__)
```
